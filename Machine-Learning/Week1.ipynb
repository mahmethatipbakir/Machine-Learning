```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = {"fiyat": [860000,820000,780000,750000,700000],
          "km":[6000,8200,9000,14200,16200]}
```

```python
dataset
```

```python
dataset = pd.DataFrame(dataset)
```

```python
dataset
```

```python
plt.plot(dataset['km'], dataset['fiyat'], marker='o')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
df = pd.DataFrame(dataset)

# Bağımsız (X) ve bağımlı (y) değişkenleri ayırma
X = df['km'].values.reshape(-1, 1)  # X: Kilometre, 2D hale getiriyoruz
y = df['fiyat']  # y: Fiyat

# Lineer regresyon modeli oluşturma
model = LinearRegression()
model.fit(X, y)

# Regresyon çizgisini oluşturmak için tahminler
y_pred = model.predict(X)

# Orijinal veri noktalarını çizme
plt.scatter(df['km'], df['fiyat'], color='blue', label='Gerçek Veriler')

# Regresyon çizgisini çizme
plt.plot(df['km'], y_pred, color='red', label='Regresyon Çizgisi')

# Grafiğe başlık ve etiketler ekleme
plt.title('Kilometreye Göre Fiyat - Doğrusal Regresyon')
plt.xlabel('Kilometre')
plt.ylabel('Fiyat')
plt.legend()

# Grafiği gösterme
plt.show()
```
