```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = {"fiyat": [860000,820000,780000,750000,700000],
          "km":[6000,8200,9000,14200,16200]}
```

```python
dataset
```

```python
dataset = pd.DataFrame(dataset)
```

```python
dataset
```

```python
plt.plot(dataset['km'], dataset['fiyat'], marker='o')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
df = pd.DataFrame(dataset)

# Bamsz (X) ve baml (y) deikenleri ayrma
X = df['km'].values.reshape(-1, 1)  # X: Kilometre, 2D hale getiriyoruz
y = df['fiyat']  # y: Fiyat

# Lineer regresyon modeli oluturma
model = LinearRegression()
model.fit(X, y)

# Regresyon izgisini oluturmak iin tahminler
y_pred = model.predict(X)

# Orijinal veri noktalarn izme
plt.scatter(df['km'], df['fiyat'], color='blue', label='Gerek Veriler')

# Regresyon izgisini izme
plt.plot(df['km'], y_pred, color='red', label='Regresyon izgisi')

# Grafie balk ve etiketler ekleme
plt.title('Kilometreye Gre Fiyat - Dorusal Regresyon')
plt.xlabel('Kilometre')
plt.ylabel('Fiyat')
plt.legend()

# Grafii gsterme
plt.show()
```
